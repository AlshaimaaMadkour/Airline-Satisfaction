```python
classes = ['Business','Eco','Eco Plus']
columns_names= ['Ease of Online booking','Gate location','Food and drink', 'Online boarding', 'Seat comfort','Inflight entertainment', 'On-board service', 'Leg room service','Baggage handling', 'Checkin service', 'Inflight service']

for class_ in classes:
    x= df[df['Class'] == class_]
    print(class_,round(len(x)/len(df)*100),'%')
    for column_name in columns_names :
        print(column_name, ':', round(x[x['Customer Type'] == 'disloyal Customer'][column_name].mean(),1))
    print(' ')
```
